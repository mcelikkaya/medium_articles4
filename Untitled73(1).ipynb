```python
import numpy as np
import matplotlib.pyplot as plt
import random
#from matplotlib.animation import FuncAnimation, PillowWriter
np.random.seed(123)

class MyKMeans:
  def __init__(self,n_cluster,n_iterations,init_method="random"):
    self.n_cluster = n_cluster
    self.n_iterations = n_iterations
    self.centroids = []
    self.init_method = init_method
    self.steps = []
    self.sum_errors = []

  def euclidean(self,point, data):
    """
    Euclidean distance between 2 points
    """
    return np.sqrt(np.sum((point - data)**2, axis=1))  
  

  def init_centroids(self,x):
    self.centroids = x[np.random.choice(len(x), self.n_cluster, replace=False), :] 
    return self.centroids     

  def calculate_new_centroids(self,x,clusters):
    new_centroids = []
    for idx in range(self.n_cluster ):
      points_in_cluster = x[clusters==idx]
      new_centroids.append( points_in_cluster.mean(axis=0))

    return np.array( new_centroids )  

  def calculate_new_clusters(self,x):    
    distances = np.array([self.euclidean(centroid, x) for centroid in self.centroids]).T    
    clusters = np.array([np.argmin(i) for i in distances]) #Step 3
    sum_error = np.sum([np.min(i) for i in distances]) #Step 3
    return clusters,sum_error      
  def predict(self,x):
    clusters,sum_error = self.calculate_new_clusters(x)
    return clusters
    
  def fit_predict(	self,x):
      self.fit(x)
      return self.predict(x)
    
    
    
  def fit(self,x):
    #Randomly choosing Centroids 
    self.steps = []
    self.sum_errors = []
    self.centroids = self.init_centroids(x)
    clusters,sum_error = self.calculate_new_clusters(x)
    self.sum_errors.append( sum_error )
    self.steps.append([clusters ,self.centroids]) 
    #Repeating the above steps for a defined number of iterations
    #Step 4
    
    
    for _ in range(self.n_iterations): 

        self.centroids = self.calculate_new_centroids(x,clusters)
         
        #distances = cdist(x, centroids ,'euclidean')
        clusters,sum_error = self.calculate_new_clusters(x)
        self.steps.append([clusters ,self.centroids])
        self.sum_errors.append( sum_error )
         
    return clusters ,self.centroids   

```

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
data = load_digits().data
pca = PCA(2)
  
#Transform the data
df = pca.fit_transform(data)
 

n_iterations = 100
n_cluster = 10

mykmeans = MyKMeans(n_cluster,n_iterations)
#Applying our function

clusters ,centroids = mykmeans.fit(df)

plt.plot(mykmeans.sum_errors,label="error")
plt.show()

u_labels = np.unique(clusters)
for i in u_labels:
    plt.scatter(df[clusters == i , 0] , df[clusters == i , 1] , label = i)
plt.plot(centroids[:,0],centroids[:,1],'k+',markersize=12)    
plt.legend()
plt.show()
```

```python
def gen_data(minx,maxx,step,noise,num_sample):
  #random centers, good for adhoc test, not good for unit tests
  #centersx = np.random.rand(step) * ( maxx - minx) + minx 
  #centersy = np.random.rand(step) * ( maxx - minx) + minx

  centersx = np.linspace(minx,maxx,step)
  centersy = np.random.rand(step) * ( maxx - minx) + minx


  clusters = []
  for center in zip(centersx,centersy):
    sx = np.random.random( (num_sample,2) ) * noise + np.array([center[0],center[1]])
    clusters.append( sx)
  for id,cluster in enumerate(clusters):
    plt.scatter(cluster[:,0] ,cluster[:,1]   ,label = str(id))
  plt.legend()  
  plt.show()  
  return np.vstack(clusters)  

def test_dataset(n_iterations,n_cluster,df,is_plot):
  mykmeans = MyKMeans(n_cluster,n_iterations)
  clusters ,centroids = mykmeans.fit(df)
  if is_plot:
    fig, axs = plt.subplots(1, 2)


    axs[0].plot(mykmeans.sum_errors,label="error")
    #plt.show()
    #Visualize the results
    u_labels = np.unique(clusters)
    for i in u_labels:
      plt.scatter(df[clusters == i , 0] , df[clusters == i , 1] , label = i)
    axs[1].plot(centroids[:,0],centroids[:,1],'k+',markersize=12)    
    axs[1].legend()
      
  return clusters ,centroids, mykmeans.steps, mykmeans.sum_errors
```
